```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('google_log.csv.zip', sep = '\t', compression='zip', parse_dates=['date'])
df['totals.transactionRevenue'] = df['totals.transactionRevenue'] / 10**6
print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize = (16,9))

sns.heatmap(df.sample(5000).isnull(),
           yticklabels = False,
           cbar = False)
```

```python
df.describe()
```

```python
df['month'] = df.date.apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['Visit_day'] = df.date.dt.day
df['Visit_month'] = df.date.dt.month

df['Visit_weekday'] = df.date.dt.weekday
df['Visit_weeknum'] = df.date.dt.weekofyear

df['Is_month_start'] = df.date.dt.is_month_start
df['Is_month_end'] = df.date.dt.is_month_end
```

```python
from datetime import datetime
```

```python
df['Visit_hours'] = df.visitStartTime.apply(lambda x: str(datetime.fromtimestamp(x).hour)).astype(int)
```

```python
df['ts'] = df.visitStartTime.apply(lambda x: pd.datetime.fromtimestamp(x))
```

### Сколько людей пользуются магазином в месяц/день?

```python
df_day = df.groupby('date')\
           .agg({'fullVisitorId':'nunique'})\
           .sort_values('date')
df_day.head()
```

```python
df_day.plot(figsize = (10,5))
plt.legend(loc = 'lower left');
plt.title(f"Average visitors per day {round(df_day.fullVisitorId.mean(),2)}")
plt.show();
```

```python
df_month = (
         df.groupby('month')\
           .agg({'fullVisitorId':'nunique'})\
           .sort_values('month')
            )
df_month.head()
```

```python
df_month.plot(figsize = (10,5))
plt.title(f"Average visitors per day {round(df_month.fullVisitorId.mean(),2)}")
plt.show();
```

### Сколько пользовательскийх сессий в день?

```python
df_session =(
    df.groupby('date')
    .agg({'fullVisitorId':'count'})
    .sort_values('date')
)

df_session.plot(figsize = (10,5))
plt.title(f"Average visitors per day {round(df_session.fullVisitorId.mean(),2)}")
plt.show();
```

### Когортный анализ

```python
first_minth_session = (
    df.groupby('fullVisitorId')
    .agg({'month':'min','date':'min','ts':'min'})
)

first_minth_session.columns = ['first_invate_month','first_invate_date','first_invate_ts']
first_minth_session.head()
```

```python
df = df.join(first_minth_session ,on= 'fullVisitorId')
```

```python
kogort_month = df[['fullVisitorId','month','first_invate_month']]
kogort_month.head()
```

```python
retention = kogort_month.pivot_table(
    index= 'first_invate_month',
    columns= 'month',
    values= 'fullVisitorId',
    aggfunc= 'nunique'
)

retention
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(retention, 
           annot= True,
           fmt= '.0f', 
           cmap= 'YlGnBu',
           linecolor = 'black',
           linewidths= 0.5, 
           vmax = 2500)
plt.show()
```

### Когда люди покупают?

```python
100*(1-df['totals.transactionRevenue'].isnull().mean())
```

```python
df_buy = df.dropna(subset=['totals.transactionRevenue'])
df_buy.shape
```

```python
fig, ax = plt.subplots(1,2, figsize = (16,6))

df_buy['totals.transactionRevenue'].hist(bins = 50, log = True, ax = ax[0])
sns.distplot(np.log(df_buy['totals.transactionRevenue']+1), ax = ax[1])

ax[0].set_title('Распределение выручки', fontsize = 20)
ax[0].set_xlabel('Выручка')

ax[1].set_title('Распределение логарифма выручки', fontsize = 20)
ax[1].set_xlabel('Логарифм выручка');
```

```python
first_day_buy = (
    df_buy.groupby('fullVisitorId')
        .agg({'month':'min',
              'date':'min',
              'ts':'min'
             })
)
```

```python
first_day_buy.columns = ['first_buy_month','first_buy_date','first_buy_ts']
first_day_buy.head()
```

```python
df_buy = df.join(first_day_buy ,on= 'fullVisitorId')
```

```python
(df_buy.first_buy_date != df_buy.first_invate_date).mean()
```

```python
df_buy['deltatime_buy'] = (df_buy['first_buy_ts'] - df_buy['first_invate_ts'])/np.timedelta64(1,'D')
df_buy['deltatime_buy'].hist(bins = 30, log = True)
plt.xlabel('Время между первой покупкой и между первым визитом');
```
